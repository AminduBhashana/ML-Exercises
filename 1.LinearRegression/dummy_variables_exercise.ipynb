```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\Data Science\\datasets\\ML\\Exercises\\1.LinearRegression\\carprices.csv")
df
```

```python
plt.figure(figsize=(14, 8))

# Scatter plot for Price vs. Mileage
plt.subplot(1, 2, 1)
plt.scatter(df['Mileage'], df['Sell Price($)'])
plt.title('Sell Price vs Mileage')
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')

# Scatter plot for Price vs. Age
plt.subplot(1, 2, 2)
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
plt.title('Sell Price vs Age')
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')

plt.tight_layout()
plt.show()
```

Here we can use Linear Regression model for the predict prices accrding to this independent variables

```python
dummies = pd.get_dummies(df['Car Model'],dtype=int)
dummies
```

```python
merged_df = pd.concat([df,dummies],axis="columns")
merged_df
```

```python
final_df = merged_df.drop(columns=["Car Model","Mercedez Benz C class"])
final_df
```

```python
model = LinearRegression()
```

```python
x = final_df.drop(columns="Sell Price($)")
x
```

```python
y = final_df['Sell Price($)']
y
```

```python
model.fit(x,y)
```

<b> Price of a merceddez benz that is 4 years old with mileage 45000</b>

```python
input_data = pd.DataFrame({
    'Mileage':[45000],
    'Age(yrs)':[4],
    'Audi A5':[0],
    'BMW X5':[0]
})
input_data
```

```python
predicted_price = model.predict(input_data)
print("Predicted Sell Price for a merceddez benz that is 4 years old with mileage 45000:",predicted_price[0],"$")
```

<b>Price of a BMW X5 that is 7 years old with mileage 86000</b>

```python
input_data = pd.DataFrame({
    'Mileage':[86000],
    'Age(yrs)':[7],
    'Audi A5':[0],
    'BMW X5':[1]
})
input_data
```

```python
predicted_price = model.predict(input_data)
print("Predicted Sell Price for a BMW X5 that is 7 years old with mileage 86000:",predicted_price[0],"$")
```

<b>Accuracy of the model</b>

```python
model.score(x,y)
```
